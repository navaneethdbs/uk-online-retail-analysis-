```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load both sheets from the excel file
df1 = pd.read_excel(r"D:\Restart job hunt\resumes\PROJECT 1\online_retail_II.xlsx",sheet_name="Year 2009-2010")
df2 = pd.read_excel(r"D:\Restart job hunt\resumes\PROJECT 1\online_retail_II.xlsx",sheet_name="Year 2010-2011")

# Combine both sheets into one dataframe
df = pd.concat([df1,df2],ignore_index = True)

print("data loaded successfully")
print("total rows:",len(df))


```

```python
# Look at the first 5 rows 

df.head()
```

```python
# Check column names and data types3
df.info()
```

```python
# Check missing values in each column
df.isnull().sum()
```

```python
# Remove rows where customer id is missing
df = df.dropna(subset=["Customer ID"])

# Confirm how many rows remains
print("Rows after removing missing Customer IDs:",len(df))
```

```python
# Remove rows where description is missing
df = df.dropna(subset=["Description"])

# Confrim how many rows remains
print("Rows after removing missing descriptions:",len(df))
```

```python
# Remove cancelled orders (invoice starting with 'c')
df = df[~df["Invoice"].astype(str).str.startswith("C")]

# Confirm how many rows remain
print("Rows after removing cancelled orders:",len(df))
```

```python
# Checking the -ve values in quantity
print(df["Quantity"].min())
```

```python
#checking the price column if zero or -ve values exists
print(df["Price"].min())
```

```python
# count rows where price is zero
print("Zero price rows:",(df["Price"]==0).sum())
```

```python
#Removing rows where price is Zero
df = df[df["Price"] > 0]

print("rows after removing zero price:",len(df))
```

```python
#fix customer id from float to integer 
df["Customer ID"] = df["Customer ID"].astype(int)

print(df["Customer ID"].dtype)
```

```python
#Create revenue column
df["Revenue"] = df["Quantity"] * df["Price"]

#check first 5 rows 
df.head()
```

```python
#Export clean data to CSV
df.to_csv(r"D:\Restart job hunt\resumes\PROJECT 1\online_retail_clean.csv",index=False)

print("Clean data exported successfully")
print("Final row count:",len(df))
```
